```python
d1 = {'a': 10, 'b': 20, 'c': 30}
d2 = {'b': 5, 'c': 15, 'd': 25}

result = d1.copy()

for key in d2:
    if key in result:
        result[key] += d2[key]
    else:
        result[key] = d2[key]

print(result)
```

```python
sentence = "apple banana apple mango banana apple"

words = sentence.split()

count = {}

for word in words:
    count[word] = words.count(word)

print(count)
```
